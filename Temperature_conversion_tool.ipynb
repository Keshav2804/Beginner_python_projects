```python
# Temperature Conversion Functions
def celsius_to_fahrenheit(celsius: float) -> float:
    """Converts Celsius to Fahrenheit."""
    return (celsius * 9/5) + 32

def fahrenheit_to_celsius(fahrenheit: float) -> float:
    """Converts Fahrenheit to Celsius."""
    return (fahrenheit - 32) * 5/9

def celsius_to_kelvin(celsius: float) -> float:
    """Converts Celsius to Kelvin."""
    return celsius + 273.15

def kelvin_to_celsius(kelvin: float) -> float:
    """Converts Kelvin to Celsius."""
    return kelvin - 273.15

def fahrenheit_to_kelvin(fahrenheit: float) -> float:
    """Converts Fahrenheit to Kelvin."""
    celsius = fahrenheit_to_celsius(fahrenheit)
    return celsius_to_kelvin(celsius)

def kelvin_to_fahrenheit(kelvin: float) -> float:
    """Converts Kelvin to Fahrenheit."""
    celsius = kelvin_to_celsius(kelvin)
    return celsius_to_fahrenheit(celsius)

# Main Program
def main():
    print("Temperature Conversion Tool")
    print("Select the conversion type:")
    print("1. Celsius to Fahrenheit")
    print("2. Fahrenheit to Celsius")
    print("3. Celsius to Kelvin")
    print("4. Kelvin to Celsius")
    print("5. Fahrenheit to Kelvin")
    print("6. Kelvin to Fahrenheit")

    while True:
        choice = input("Enter your choice (1-6) or 'q' to quit: ")
        
        if choice == 'q':
            break

        if choice in ['1', '2', '3', '4', '5', '6']:
            temp = float(input("Enter the temperature value: "))

            if choice == '1':
                print("{}°C is {:.2f}°F".format(temp, celsius_to_fahrenheit(temp)))
            elif choice == '2':
                print("{}°F is {:.2f}°C".format(temp, fahrenheit_to_celsius(temp)))
            elif choice == '3':
                print("{}°C is {:.2f}K".format(temp, celsius_to_kelvin(temp)))
            elif choice == '4':
                print("{}K is {:.2f}°C".format(temp, kelvin_to_celsius(temp)))
            elif choice == '5':
                print("{}°F is {:.2f}K".format(temp, fahrenheit_to_kelvin(temp)))
            elif choice == '6':
                print("{}K is {:.2f}°F".format(temp, kelvin_to_fahrenheit(temp)))

            # Optional: Ask if the user wants to perform another conversion
            another_conversion = input("Do you want to perform another conversion? (y/n): ")
            if another_conversion.lower() != 'y':
                break
        else:
            print("Invalid choice. Please enter a number between 1 and 6.")

if __name__ == "__main__":
    main()
```
