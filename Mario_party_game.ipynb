```python
import random

class Player:
    def __init__(self, name):
        self.name = name
        self.position = 0
        self.coins = 0
    
    def move(self, steps):
        self.position += steps
        self.position %= 20  # Assuming a board with 20 spaces
    
    def add_coins(self, amount):
        self.coins += amount
    
    def __str__(self):
        return f"{self.name} is at position {self.position} with {self.coins} coins."

def roll_dice():
    return random.randint(1, 6)

def random_event(player):
    event = random.choice(['gain', 'lose', 'nothing'])
    if event == 'gain':
        coins = random.randint(1, 10)
        player.add_coins(coins)
        print(f"{player.name} found {coins} coins!")
    elif event == 'lose':
        coins = random.randint(1, 5)
        player.add_coins(-coins)
        print(f"{player.name} lost {coins} coins!")
    else:
        print(f"{player.name} encountered nothing special.")

def take_turn(player):
    print(f"\n{player.name}'s turn:")
    steps = roll_dice()
    print(f"{player.name} rolled a {steps}.")
    player.move(steps)
    print(player)
    random_event(player)

def main():
    num_players = int(input("Enter the number of players: "))
    players = [Player(input(f"Enter name for player {i+1}: ")) for i in range(num_players)]
    
    turns = 10
    for turn in range(turns):
        print(f"\n--- Turn {turn + 1} ---")
        for player in players:
            take_turn(player)
    
    print("\n--- Final Results ---")
    for player in players:
        print(player)

    winner = max(players, key=lambda p: p.coins)
    print(f"\nCongratulations, {winner.name} wins with {winner.coins} coins!")

if __name__ == "__main__":
    main()
```
